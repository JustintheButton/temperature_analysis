```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\hhuan\TemparatureStudy\vertical-jetty-291800-46b4a02fef85.json"
from google.cloud import bigquery
import pandas
import pandas as pd
import matplotlib.pyplot as plt
client = bigquery.Client()
```

```python
useful_stations={}
for year in range(1930, 2022):
    sql_command =     """
        SELECT stn, wban, count(1) as num
        FROM `bigquery-public-data.noaa_gsod.gsod{}` 
        group by stn, wban
        """.format(str(year))
    query_job = client.query(sql_command)
    results = query_job.result().to_dataframe()
    useful_stations[year] = results[(results['num']>360) & (results['num']<380)]
    print(year)
```

```python
useful_sattions_play = {}
for key in useful_stations:
    useful_sattions_play[key] = useful_stations[key].set_index(['stn', 'wban'])
```

```python
stations={}
station_set=set({})
for start_year in range(1930, 1971):
    stations_list = useful_sattions_play[start_year].index.tolist()
    count = {}
    for a_station in stations_list:
        count[a_station] = 0
        for key in useful_sattions_play:
            if a_station in useful_sattions_play[key].index:
                count[a_station] += 1

    stations[start_year]=[]
    for key in count:
        if (count[key] > 2021-start_year-5) and (key[0] not in station_set):
            stations[start_year].append(key[0])
            station_set.add(key[0])

for key in stations:
    print(key, stations[key])
```
