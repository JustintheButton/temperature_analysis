```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\hhuan\TemparatureStudy\vertical-jetty-291800-46b4a02fef85.json"
from google.cloud import bigquery
import pandas
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
client = bigquery.Client()
```

```python
import json
with open('new_stations_by_year.txt') as f:
    data = f.read()
stations=json.loads(data)
```

```python
print(stations)
```

```python
for start_year in range(1950, 1971):
    new_stations_found_in_this_year = stations[str(start_year)]
    for station_id in new_stations_found_in_this_year:
        for a_year in range(start_year, 2021):
            file_name = 'data/'+str(station_id)+'_'+str(a_year)+'.csv'
            path = Path(file_name)
            if path.is_file():
                continue
            if Path('data_wholeyear/'+str(station_id)+'.csv').is_file():
                continue
            sql_command =     """
                SELECT stn, year, mo, da, temp, max, min
                FROM `bigquery-public-data.noaa_gsod.gsod{}` 
                where stn='{}'
                """.format(str(a_year),str(station_id))
            query_job = client.query(sql_command)
            results = query_job.result().to_dataframe()
            print(station_id, a_year, len(results))
            if len(results)>0:
                results.to_csv('data/'+str(station_id)+'_'+str(a_year)+'.csv')
```
